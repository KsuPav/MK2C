**База**

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

data = pd.read_csv('/content/drive/My Drive/ШКОЛА ММ/data.csv', encoding = 'utf-8')

data = pd.DataFrame(data)
```

```python
data = data.dropna()
data = data.convert_dtypes()
```

```python
# вычисление общего времени (пробки?)
data['Общее_время'] = [2*x for x in data['Удаленность']]
data['Общее_время'] = data['Общее_время'] + data['Время_на_посещение']
```

```python
data
```

**Удаление мест, которые не подходят по времени**

```python
# информация о билете
a = int(input())
b = int(input())
```

```python
# удаляем то, что не подходит по времени
data_time = data
for i in range(data_time.count()[0]):

  a_plus_road = a + data_time.loc[i]['Удаленность']
  b_minus_road = b - data_time.loc[i]['Удаленность']
  
  if a_plus_road > 24: a_plus_road -= 24
  if b_minus_road < 0: b_minus_road += 24
  
  dia_left = max(a_plus_road, data_time.loc[i]['Время_С'])
  dia_right = min(b_minus_road, data_time.loc[i]['Время_ДО'])
  dia = dia_right - dia_left
  
  if dia < data_time.loc[i]['Время_на_посещение']:
    data_time = data_time.drop(labels = [i], axis = 0)

data_time = data_time.reset_index(drop=True)
```

**Сортировка по тегам**

```python
place = 'Улица' # Помещение / Улица
activity = 'Экскурсия' # Экскурсия / Прогулка / Питание / Сон / Кино
```

```python
from sklearn import preprocessing
import numpy as np

norm = np.array(data_time['Удаленность'].tolist())
norm = preprocessing.normalize([norm])
norm = np.round(norm, 2)[0]

data_pri = data_time
data_pri['Удаленность_норм'] = norm
```

```python
from sklearn import preprocessing

data_pri['Приоритет'] = [0]*len(data_pri.index)
data_pri['Доп'] = [0]*len(data_pri.index)

data_pri['Доп'] = [1 if x == place else 0 for x in data_pri['Место']]
data_pri['Приоритет'] = data_pri['Приоритет'] + data_pri['Доп']

data_pri['Доп'] = [1 if x == activity else 0 for x in data_pri['Активность']]
data_pri['Приоритет'] = data_pri['Приоритет'] + data_pri['Доп']

data_pri['Приоритет'] = data_pri['Приоритет'] + (1 - data_pri['Удаленность_норм'])

data_pri.pop('Доп')
#data_pri.pop('Удаленность_норм')


data_pri
```

```python
data_sorted_pri = data_pri.sort_values(by='Приоритет', ascending=False, ignore_index=True)

data_sorted_pri
```
