```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

data = pd.read_csv('/content/drive/My Drive/ШКОЛА ММ/data.csv', encoding = 'utf-8')

data = pd.DataFrame(data)
```

```python
data
```

```python
data.pop('Город')
data.pop('Аэропорт')
data.pop('Можно с детьми')
```

```python
import random

dost = list(data['Достопремечательность'])
data_new = pd.DataFrame()

for i in range(1,11):
  for j in dost:
    new_row = {'Юзер':i,
               'Достопремечательность':j,
               'Оценка':random.choice([' ', 1, 2, 3, 4, 5])}
    data_new = data_new.append(new_row, ignore_index=True)
```

```python
data_new
```

```python
full_data = pd.merge(data_new, data, on='Достопремечательность')
```

```python
full_data.convert_dtypes() 
full_data
```

```python
full_data = full_data.loc[full_data['Оценка'] != ' ']
full_data['Оценка'] = full_data['Оценка'].astype(int)
full_data['Юзер'] = full_data['Юзер'].astype(int)
```

```python
full_data.dtypes
```

```python
# Средний рейтинг каждого места
full_data.groupby('Достопремечательность')['Оценка'].mean()
```

```python
# Отсортированный средний рейтинг каждого места
full_data.groupby('Достопремечательность')['Оценка'].mean().sort_values(ascending=False)
```

```python
# Общее кол-во оценок для каждого места
full_data.groupby('Достопремечательность')['Оценка'].count().sort_values(ascending=False)
```

```python
# Матрица названий фильмов и соответствующих пользовательских оценок
Name_User_Rat = full_data.pivot_table(index='Юзер', columns='Достопремечательность', values='Оценка')
Name_User_Rat 
```
